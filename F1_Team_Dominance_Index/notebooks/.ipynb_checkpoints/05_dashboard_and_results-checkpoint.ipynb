# üèÅ Dashboard & Final Results

This notebook summarizes the Team Dominance Index (TDI) results for Formula 1 (1950‚Äì2025).  
It loads the final processed TDI dataset and produces leaderboards, era comparisons, and exportable outputs.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import os
pd.set_option('display.max_columns', None)
pd.set_option('display.precision', 6)

base_dir = r"C:\Users\ASUS\OneDrive\Documents\Desktop\F1_Dominance_Analytics\F1_Team_Dominance_Index"
results_path = os.path.join(base_dir, "output", "results", "final_team_tdi.csv")

if not os.path.exists(results_path):
    raise FileNotFoundError(f"Final TDI file not found at: {results_path}")

tdi_df = pd.read_csv(results_path)

print(" Final TDI Data Loaded Successfully!")
print(f"Shape: {tdi_df.shape}")
print("\nColumns:", list(tdi_df.columns))
display(tdi_df.head(10))
```

## Top Teams & Most Dominant Seasons

This section identifies:

- **Top 10 Most Dominant Seasons in F1 history**
- **Top 10 Teams by Average TDI across all seasons**
- **Each team‚Äôs Best-Ever Dominant Season**
- Visualization: Top 10 dominant seasons with a horizontal bar chart, highlighting which team-season combinations were most dominant.

```python
top_seasons = (
    tdi_df.sort_values(by="TDI", ascending=False)
    .head(10)[["year", "name", "TDI"]]
    .reset_index(drop=True)
)

print("üî• Top 10 Most Dominant Seasons in F1 History:")
display(top_seasons)

# --- Top 10 Teams by Average TDI ---
avg_tdi = (
    tdi_df.groupby("name", as_index=False)["TDI"]
    .mean()
    .sort_values(by="TDI", ascending=False)
    .head(10)
)
print("\nüèÜ Top 10 Teams by Average Dominance (Across All Seasons):")
display(avg_tdi)

# --- Each Team's Best Season ---
best_per_team = (
    tdi_df.loc[tdi_df.groupby("name")["TDI"].idxmax()][["name", "year", "TDI"]]
    .sort_values(by="TDI", ascending=False)
    .reset_index(drop=True)
)
print("\n‚≠ê Each Team‚Äôs Best-Ever Dominant Season:")
display(best_per_team.head(10))

# --- Visualization: Top 10 Dominant Seasons ---
fig = px.bar(
    top_seasons,
    x="TDI",
    y=top_seasons["year"].astype(str) + " - " + top_seasons["name"],
    orientation="h",
    title="üî• Top 10 Most Dominant Seasons in F1 History (TDI)",
    color="TDI",
    color_continuous_scale="Reds",
)
fig.update_layout(
    xaxis_title="Team Dominance Index (TDI)",
    yaxis_title="Season",
    template="plotly_dark",
    height=500
)
fig.show()
```


## Dominance Evolution Over Time

- Explore how the TDI of major teams evolved from 1950‚Äì2025.
- Trend lines illustrate era-defining streaks ‚Äî for example, Ferrari‚Äôs early 2000s, Mercedes‚Äô turbo-hybrid dominance, Red Bull‚Äôs rise after 2010.
- Calculate each team‚Äôs average TDI across all seasons to quantify long-term consistency.

**Visualization:**  
Line graph showing TDI trends by team across decades.

```python
# -----------------------------------------------
# üèéÔ∏è STEP 3 ‚Äî Dominance Trends Over Time
# -----------------------------------------------

# Focus on top historical constructors
top_teams_focus = ["Ferrari", "McLaren", "Mercedes", "Red Bull", "Williams", "Renault"]

# Filter dataset for these teams only
trend_df = tdi_df[tdi_df["name"].isin(top_teams_focus)].copy()

# Sort by year for smooth plotting
trend_df = trend_df.sort_values(by=["name", "year"])

# --- Interactive Line Chart: TDI over Time ---
fig = px.line(
    trend_df,
    x="year",
    y="TDI",
    color="name",
    title="üèéÔ∏è Dominance Evolution (1950‚Äì2025): TDI Trend by Team",
    markers=True,
)

fig.update_traces(mode="lines+markers", line=dict(width=2))
fig.update_layout(
    xaxis_title="Season",
    yaxis_title="Team Dominance Index (TDI)",
    legend_title="Constructor",
    template="plotly_dark",
    height=600,
    font=dict(size=13, color="white"),
    plot_bgcolor="#111111",
    paper_bgcolor="#111111",
)

fig.show()

# --- Summary Insight Table: Average Dominance ---
dominance_summary = (
    trend_df.groupby("name")["TDI"]
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

dominance_summary["TDI"] = dominance_summary["TDI"].round(3)

print("üìä Average Team Dominance (1950‚Äì2025):")
display(dominance_summary)
```


## Era-wise (Decade) Dominance Analysis

- For every decade, compute the most dominant team by averaging TDI values.
- Reveal which constructor defined each era ‚Äî from Alfa Romeo in the 1950s to Red Bull in the 2020s.


```python
tdi_df["decade"] = (tdi_df["year"] // 10) * 10

decade_avg = (
    tdi_df.groupby(["decade", "name"], as_index=False)["TDI"]
    .mean()
)

top_decade_team = (
    decade_avg.loc[decade_avg.groupby("decade")["TDI"].idxmax()]
    .sort_values(by="decade")
    .reset_index(drop=True)
)

print("üèÅ Era-wise Dominance Leaders:")
display(top_decade_team)

fig = px.bar(
    top_decade_team,
    x="decade",
    y="TDI",
    color="name",
    text="name",
    title="üèÜ Decade-wise F1 Dominance (Highest Avg TDI per Decade)",
    color_discrete_sequence=px.colors.qualitative.Bold,
)

fig.update_traces(textposition="outside")
fig.update_layout(
    xaxis_title="Decade",
    yaxis_title="Average Team Dominance Index (TDI)",
    template="plotly_dark",
    height=500,
    font=dict(size=13, color="white"),
    plot_bgcolor="#111111",
    paper_bgcolor="#111111",
)
fig.show()
```

```python
# ‚úÖ Step 4.5 ‚Äî Decade-wise Dominance Summary

# Create a new column for decade (e.g., 1950s, 1960s, etc.)
tdi_df["decade"] = (tdi_df["year"] // 10) * 10

# Find the top team per decade (highest average TDI within that decade)
decade_leaders = (
    tdi_df.groupby(["decade", "name"], as_index=False)["TDI"]
    .mean()
    .sort_values(["decade", "TDI"], ascending=[True, False])
    .groupby("decade")
    .head(1)
    .reset_index(drop=True)
)

print("üèÅ Decade-Wise Most Dominant Teams:")
display(decade_leaders)
```

## Export Dashboard Summary
- Export all summarized insights ‚Äî including top decade leaders ‚Äî into

```python
import os
os.makedirs("output/results", exist_ok=True)

dominance_summary.to_csv("output/results/average_tdi_by_team.csv", index=False)
top_seasons.to_csv("output/results/top_10_dominant_seasons.csv", index=False)
decade_leaders.to_csv("output/results/decade_dominance_summary.csv", index=False)

print("Dashboard summary files exported successfully:")
print(" - output/results/average_tdi_by_team.csv")
print(" - output/results/top_10_dominant_seasons.csv")
print(" - output/results/decade_dominance_summary.csv")
```
