This Venn diagram shows the overlap between **wins**, **podiums**, and **other finishes** for **Brawn GP** in the 2009 Formula 1 season.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

results = pd.read_csv("C:/Users/ASUS/OneDrive/Documents/Desktop/F1_Dominance_Analytics/F1_Team_Dominance_Index/data/raw/f1db-races-race-results.csv")

brawn_2009 = results[(results['constructorId'] == 'brawn') & (results['year'] == 2009)]

total_races = brawn_2009['raceId'].nunique()
wins = brawn_2009[brawn_2009['positionNumber'] == 1]['raceId'].nunique()
podiums = brawn_2009[brawn_2009['positionNumber'].isin([1, 2, 3])]['raceId'].nunique()

only_wins = wins 
only_podiums = podiums - wins 
others = total_races - podiums 

plt.figure(figsize=(7,7))
venn3(
    subsets=(only_wins, only_podiums, others, 0, 0, 0, 0),
    set_labels=('Wins', 'Podiums', 'Other Races'),
    set_colors=('gold', 'lightblue', 'lightgrey'),
    alpha=0.7
)
plt.title("Brawn GP (2009): Race Distribution (17 GPs)")
plt.text(-0.8, 0.6, f"Wins: {wins}", fontsize=11)
plt.text(0.4, 0.6, f"Podiums: {podiums}", fontsize=11)
plt.text(0.0, -0.7, f"Total Races: {total_races}", fontsize=11)
plt.show()

print(f"üèÜ Total Races: {total_races}")
print(f"ü•á Wins: {wins}")
print(f"ü•à Podiums (including wins): {podiums}")
print(f"‚ö™ Other Races (no podium): {others}")
```

### üèÜ Dominance Summary Table

This table lists the top constructors with the **most dominant seasons** in F1 history (TDI > 0.75).  
It also shows each team‚Äôs **peak season** ‚Äî the year they achieved their highest Team Dominance Index (TDI) ‚Äî along with the corresponding **peak TDI value**.


```python
dominance_threshold = 0.75

dominance_summary = (
    df.groupby("name")
    .agg(
        dominant_years = ("TDI", lambda x: (x > dominance_threshold).sum()),
        peak_year = ("year", lambda x: df.loc[x.index, "TDI"].idxmax()),
        peak_TDI = ("TDI", "max")
    )
    .reset_index()
)

dominance_summary["peak_year"] = dominance_summary["peak_year"].apply(
    lambda i: df.loc[i, "year"] if i in df.index else None
)

dominance_summary = dominance_summary.sort_values(by="dominant_years", ascending=False)
print("‚úÖ Dominance duration summary created")
dominance_summary.head(10)
```

>Ferrari leads with an unmatched 24 dominant seasons, followed by McLaren, Williams, Mercedes, and Red Bull ‚Äî showcasing the dynasties that defined F1‚Äôs competitive eras.

```python
import pandas as pd
from pathlib import Path
import numpy as np

base = Path.cwd()
if base.name == "notebooks":
    base = base.parent

proc_dir = base / "data" / "processed"
infile = None
# Prefer team_year_summary if present
for name in ["team_year_summary.csv", "team_dominance_index_with_games.csv", "team_dominance_index.csv"]:
    p = proc_dir / name
    if p.exists():
        infile = p
        break

if infile is None:
    raise FileNotFoundError("No processed team summary file found; please run Step 1 inspection and share filenames.")

df = pd.read_csv(infile)
print("Loaded:", infile.name, "shape:", df.shape)

# Identify races column (races / total_races) and team name columns
if "races" in df.columns:
    df["total_races"] = df["races"]
elif "total_races" in df.columns:
    df["total_races"] = df["total_races"]
elif "races_played" in df.columns:
    df["total_races"] = df["races_played"]
else:
    # fallback: if there is a column 'round' or 'scheduled_laps' not useful; set to 0 and warn
    df["total_races"] = df.get("total_games", np.nan)  # might be present
    print("‚ö†Ô∏è Could not find 'races' or 'total_races' column ‚Äî 'total_races' filled from 'total_games' or left as NaN.")

# Ensure wins/podiums/total_points exist
for col in ["wins", "podiums", "total_points"]:
    if col not in df.columns:
        print(f"‚ö†Ô∏è Column '{col}' missing ‚Äî creating with zeros.")
        df[col] = 0

# Compute base rates safely (avoid divide by zero)
df["win_rate"] = df["wins"] / df["total_races"].replace({0: pd.NA})
df["podium_rate"] = df["podiums"] / df["total_races"].replace({0: pd.NA})

# points_share: per year fraction of total points (if total_points present)
if "total_points" in df.columns and df["total_points"].notna().any():
    df["points_share"] = df["total_points"] / df.groupby("year")["total_points"].transform("sum")
else:
    df["points_share"] = 0.0

# one_two_rate: best-effort fallback if exact one-two data not present
if "one_two_finishes" in df.columns:
    df["one_two_rate"] = df["one_two_finishes"] / df["total_races"].replace({0: pd.NA})
else:
    # As a conservative proxy, estimate one_two_rate from wins & podiums:
    # teams with many wins likely had 1-2s ‚Äî use wins/total_races scaled down
    df["one_two_rate"] = (df["wins"] / df["total_races"].replace({0: pd.NA})) * 0.25
    # note: this is a proxy; to compute exact 1-2 finishes you'd need per-race results.

# Replace inf / NaN with 0 for safety in outputs
df[["win_rate","podium_rate","points_share","one_two_rate"]] = df[["win_rate","podium_rate","points_share","one_two_rate"]].fillna(0).replace([np.inf, -np.inf], 0)

outpath = proc_dir / "team_year_summary_with_metrics.csv"
df.to_csv(outpath, index=False)
print("Saved metrics-enhanced file to:", outpath)
display(df.head(5))
```
