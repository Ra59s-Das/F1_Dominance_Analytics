### 🏁 Why Brawn GP Tops the Dominance Chart

Brawn GP’s presence at the top of the Team Dominance Index (TDI) chart is a fascinating anomaly in Formula 1 history.  
Unlike long-standing giants such as Mercedes or Ferrari, **Brawn GP competed for only one season — 2009** — yet achieved dominance levels rarely seen in the sport.

#### 🔹 The 2009 Season Miracle

- Entered the grid after **Honda’s sudden exit**, with minimal resources and limited preseason testing.  
- The team’s **double diffuser innovation** gave them a massive aerodynamic edge early in the season.  
- **Jenson Button** won **6 of the first 7 races**, and Brawn GP ultimately claimed **8 victories** out of **17 Grands Prix**.  
- They secured **both the Drivers’ and Constructors’ Championships** — an unprecedented achievement for a debut team.  

#### 📊 Why Their TDI Is So High

The TDI calculation combines win rate, podium rate, points share, and 1–2 finishes.  
Because Brawn GP:
- Won **47% of all races in 2009**,  
- Took **15 total podiums**, and  
- Outperformed all rivals in both qualifying and race pace,  

their single-season metrics heavily outweigh multi-year averages of other teams.  
This results in **an exceptionally high mean TDI**, placing them at the **top of the dominance bar graph**, despite their brief presence in the sport.

#### 🏆 Legacy

Brawn GP’s one-season triumph remains one of Formula 1’s most legendary stories — a perfect example of *short-lived but absolute dominance*.  
It stands as statistical proof that a team doesn’t need longevity to achieve greatness — just one flawless year of engineering brilliance and execution.

### 🏁 Brawn GP’s 2009 Dominance — Venn Diagram View

This Venn diagram shows the overlap between **wins**, **podiums**, and **other finishes** for **Brawn GP** in the 2009 Formula 1 season.

- **Wins (8 races):** Brawn GP won nearly **half of the 17 races**, an extraordinary single-season record.  
- **Podiums (15 total):** Even when not winning, they almost always finished in the **top 3**.  
- **Intersection:** Every win was also a podium, highlighting their **consistency and dominance**.  
- **Outside Area:** Only a couple of races saw them **off the podium**, showing rare underperformance.

📊 **Insight:**  
Brawn GP’s 2009 season was defined by unmatched efficiency — **high win rate, consistent podiums**, and near-perfect execution, which explains why their **TDI score ranks the highest** in our analysis.


```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

results = pd.read_csv("C:/Users/ASUS/OneDrive/Documents/Desktop/F1_Dominance_Analytics/F1_Team_Dominance_Index/data/raw/f1db-races-race-results.csv")

brawn_2009 = results[(results['constructorId'] == 'brawn') & (results['year'] == 2009)]

total_races = brawn_2009['raceId'].nunique()
wins = brawn_2009[brawn_2009['positionNumber'] == 1]['raceId'].nunique()
podiums = brawn_2009[brawn_2009['positionNumber'].isin([1, 2, 3])]['raceId'].nunique()

only_wins = wins 
only_podiums = podiums - wins 
others = total_races - podiums 

plt.figure(figsize=(7,7))
venn3(
    subsets=(only_wins, only_podiums, others, 0, 0, 0, 0),
    set_labels=('Wins', 'Podiums', 'Other Races'),
    set_colors=('gold', 'lightblue', 'lightgrey'),
    alpha=0.7
)
plt.title("Brawn GP (2009): Race Distribution (17 GPs)")
plt.text(-0.8, 0.6, f"Wins: {wins}", fontsize=11)
plt.text(0.4, 0.6, f"Podiums: {podiums}", fontsize=11)
plt.text(0.0, -0.7, f"Total Races: {total_races}", fontsize=11)
plt.show()

print(f"🏆 Total Races: {total_races}")
print(f"🥇 Wins: {wins}")
print(f"🥈 Podiums (including wins): {podiums}")
print(f"⚪ Other Races (no podium): {others}")
```

## 📈 Dominance Trend Over Time (1950–2025)

This line plot shows how the **Team Dominance Index (TDI)** evolved for the five most dominant F1 teams —  
**Ferrari, Mercedes, Red Bull, McLaren, and Williams** — across different eras.

It highlights iconic periods like **Ferrari’s early 2000s run**, **Mercedes’ hybrid-era control**, and **Red Bull’s recent streak**.

> ⚪ **Note — Brawn GP (2009):**  
> Brawn isn’t shown here since they raced **only one season (2009)**, but their TDI that year was **extraordinarily high**,  
> making it one of the **most dominant single-year performances** in F1 history.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load processed Team Dominance Index data
team_tdi_df = pd.read_csv(
    "C:/Users/ASUS/OneDrive/Documents/Desktop/F1_Dominance_Analytics/F1_Team_Dominance_Index/data/processed/team_dominance_index.csv"
)

# Define the top teams for visualization
top_teams_list = ["Ferrari", "Mercedes", "Red Bull", "McLaren", "Williams"]

plt.figure(figsize=(12,6))

# Plot TDI trend for each team
for team in top_teams_list:
    team_data = team_tdi_df[team_tdi_df["name"].str.contains(team, case=False, na=False)]
    plt.plot(team_data["year"], team_data["TDI"], marker='o', linewidth=2, label=team)

# Add titles, labels, and styling
plt.title("📈 F1 Team Dominance Trends (1950–2025)", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Team Dominance Index (TDI)", fontsize=12)
plt.legend(title="Teams", fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the processed dataset
df = pd.read_csv("../data/processed/team_dominance_index.csv")

# Find each team's single best season (max TDI)
peak_df = df.loc[df.groupby("name")["TDI"].idxmax()]

# Sort by TDI descending, take top 15 for clarity
top_peaks = peak_df.sort_values("TDI", ascending=False).head(15)

plt.figure(figsize=(10,6))
bars = plt.barh(top_peaks["name"], top_peaks["TDI"], color="skyblue", edgecolor="black")
plt.gca().invert_yaxis()

# Highlight Brawn GP if present
for bar, team in zip(bars, top_peaks["name"]):
    if "brawn" in team.lower():
        bar.set_color("gold")
        bar.set_edgecolor("black")

# Label each bar with year and TDI
for i, (tdi, year) in enumerate(zip(top_peaks["TDI"], top_peaks["year"])):
    plt.text(tdi + 0.01, i, f"{year}  ({tdi:.2f})", va="center", fontsize=9)

plt.title("🏁 Peak Dominance by Team (Best Season per Constructor)", fontsize=14, fontweight="bold")
plt.xlabel("TDI (Team Dominance Index)")
plt.ylabel("Constructor")
plt.grid(axis="x", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Define threshold for "dominant" season
dominance_threshold = 0.75

# Count dominant years and find peak TDI year
dominance_summary = (
    df.groupby("name")
    .agg(
        dominant_years = ("TDI", lambda x: (x > dominance_threshold).sum()),
        peak_year = ("year", lambda x: df.loc[x.index, "TDI"].idxmax()),
        peak_TDI = ("TDI", "max")
    )
    .reset_index()
)

# Convert peak_year index to actual year value
dominance_summary["peak_year"] = dominance_summary["peak_year"].apply(
    lambda i: df.loc[i, "year"] if i in df.index else None
)

# Sort by number of dominant years
dominance_summary = dominance_summary.sort_values(by="dominant_years", ascending=False)
print("✅ Dominance duration summary created")
dominance_summary.head(10)
```

```python
import matplotlib.pyplot as plt

# Define updated team colors
team_colors = {
    "McLaren": "#FF6F00",      # Deep McLaren orange
    "Red Bull": "#1A0033",     # Deep blackish blue-purple (Red Bull navy tone)
    "Mercedes": "#0D0D0D",     # Matte blackish grey
    "Cooper": "#003B2E",       # Classic dark green
    "Ferrari": "#C00000"       # Rich crimson red
}

# Select top 5 peak seasons by wins
peak_seasons = (
    team_tdi_df.loc[team_tdi_df.groupby("name")["wins"].idxmax(),
                    ["name", "year", "wins", "podiums", "races"]]
    .sort_values("wins", ascending=False)
    .reset_index(drop=True)
)

top5_peak = peak_seasons.head(5)
colors = [team_colors.get(team, "#808080") for team in top5_peak["name"]]

# Plot
plt.figure(figsize=(10,6))
bars = plt.bar(top5_peak["name"], top5_peak["wins"], color=colors, edgecolor='black', linewidth=1.2)

# Annotate
for i, row in top5_peak.iterrows():
    plt.text(i, row["wins"] + 0.3, 
             f"{row['year']} ({int(row['podiums'])}P/{int(row['races'])}R)", 
             ha='center', fontsize=9, fontweight='bold', color='black')

# Titles and styling
plt.title("🏁 Top 5 Most Dominant F1 Seasons (by Wins)", fontsize=14, weight='bold')
plt.ylabel("Total Wins", fontsize=11)
plt.xlabel("Team", fontsize=11)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

top5_peak
```

### Season-wise Dominance Comparison

In this visualization, we compare how dominant each constructor was in a specific **F1 season** based on their **Team Dominance Index (TDI)**.  

By selecting any year, we can instantly see which teams performed the strongest and how the competitive balance shifted across eras.  

Each bar represents a team’s **TDI value**, combining their performance metrics — including **win rate**, **podium share**, **average finish**, and **total points contribution** — into a single dominance score.  

Color coding aligns with official team colors (Ferrari Red, McLaren Papaya, Red Bull Blue, etc.) to make the chart both accurate and visually familiar to F1 fans.  

> 💡 **What is a Constructor?**                                                                                                               
> In Formula 1, a *constructor* refers to the team or manufacturer that designs and builds the car’s chassis and competes in the championship.  
> Examples include **Ferrari**, **McLaren**, **Mercedes**, and **Red Bull Racing**.                                                            
> Constructors accumulate points throughout the season to compete for the **Constructors’ Championship**, separate from the Drivers’ title.    


```python

import matplotlib.pyplot as plt
import seaborn as sns

selected_year = 2023

year_df = df[df["year"] == selected_year].sort_values("TDI", ascending=False)

team_colors = {
    "Ferrari": "#C00000",     
    "Red Bull": "#1A0033",     
    "Mercedes": "#0D0D0D",    
    "McLaren": "#FF8700",     
    "Aston Martin": "#00594F", 
    "Williams": "#005AFF",    
    "Alpine": "#0090FF",      
    "AlphaTauri": "#2B4562",   
    "Alfa Romeo": "#900000",   
    "Haas": "#B6BABD"          
}

colors = [team_colors.get(name, "grey") for name in year_df["name"]]

plt.figure(figsize=(10, 6))
sns.barplot(
    x="TDI",
    y="name",
    data=year_df,
    palette=colors
)

# Add text labels
for index, value in enumerate(year_df["TDI"]):
    plt.text(value + 0.005, index, f"{value:.3f}", va='center')

plt.title(f"🏆 Team Dominance Index by Constructor – {selected_year}", fontsize=15, weight='bold', pad=15)
plt.xlabel("Team Dominance Index (TDI)")
plt.ylabel("Constructor")
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
